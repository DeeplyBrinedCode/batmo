Polar graphing setup. Imports are below.

```python
import numpy as np
import pandas as pd
import matplotlib
import math
```

Now for graph setup. The data points we read will give us an angle and a distance. An example data point is below.

```python
new_angle_data = 40
new_position_data = 0.4
new_data_polar = (new_angle_data, new_position_data)
```

We'll want to turn this data into a cartesian point.

```python
new_data_cartesian = (new_data_polar[1]*math.cos(new_data_polar[0]),new_data_polar[1]*math.sin(new_data_polar[0]))
```

Now we can plot it on a graph.

```python
import matplotlib.pyplot as plt

plt.style.use('_mpl-gallery')

plt.xlabel = 'Horizontal Position'
plt.ylabel = 'Vertical Position'

x = new_data_cartesian[0]
y = new_data_cartesian[1]

# plot
fig, ax = plt.subplots()

ax.scatter(x, y)

ax.set(xlim=(-1, 1), xticks=np.arange(-1.0, 1.1, 0.5),
       ylim=(-1, 1), yticks=np.arange(-1, 1.1, 0.5), title='Position Data',
       xlabel = 'Horizontal Position', ylabel = 'Vertical Position')

plt.show()
```
